# Density and Contour Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## Visualizing a Three-Dimensional Function

```python
# definition of z = f(x, y)
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# three arguments: a grid of x values, a grid of y values, and a grid of z values
# negative values are represented by dashed lines
plt.contour(X, Y, Z, colors='black');
```

```python
# more lines to be drawn—20 equally spaced intervals within the data range
# RdGy (short for Red-Gray) colormap,
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
# to get more color impact, use contourf (f for filled contour)
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
# the plt.imshow() function, which interprets a two-dimensional grid of data as an image
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
# imshow doesn't accept an x and y grid, so you must manually specify the extent [xmin, xmax, ymin, ymax] of the image on the plot.
plt.colorbar()
plt.axis(aspect='image');
```

```python
# another way to display data with imshow and clabel
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
