# Simple Line Plots

```python
# creating symple y = f(x) function vizualisation
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# figure contains all the objects representing axes, graphics, text, and labels
fig = plt.figure()

# axes is what we see above: a bounding box with ticks and labels
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x)); # use ax for ploting
```

```python
# use of pylab interface
plt.plot(x, np.sin(x));
```

```python
# plot function multiple times
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

## Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
# Matplotlib will automatically cycle through a set of default colors for multiple lines
# linestyle keyword
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
# single non keyword argument
# standard abbreviations in the RGB (Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## Adjusting the Plot: Axes Limits

```python
# axes limits
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# reverse the order of the arguments
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
# [xmin, xmax, ymin, ymax]
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
# tighten the bounds around the current plot
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
# one unit in x is equal to one unit in y
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## Labeling Plots

```python
# set labels
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend() # create legend
```

## Aside: Matplotlib Gotchas

```python
# using object-oriented interface to plotting
ax = plt.axes()
ax.plot(x, np.sin(x), label='sin(x)')
# need to use set or set_xxxxx
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
ax.legend()
```
