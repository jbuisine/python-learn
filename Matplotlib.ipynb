# Matplotlib use

## Imports

```python
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
from matplotlib import gridspec
import pandas as pd
import random
import numpy as np
```

## Display image

```python
img = mpimg.imread('http://www.cs.us.es/~fsancho/images/2017-09/ia-imagen.png')
img.shape
```

```python
plt.imshow(img)
plt.show()
```

## Reshape image

```python
reshapeImg = img.reshape(1000, 640, 3)
plt.imshow(reshapeImg)
plt.show()
```

## Create RGB image

```python
def generate_random_pixel(a, b):
    return [random.randint(a, b) for _ in range(3)]
    
img = np.array([generate_random_pixel(50, 240) for _ in range(400)])

print(img)
img = img.reshape(20, 20, 3)
img.shape
plt.imshow(img)
plt.show()
```
